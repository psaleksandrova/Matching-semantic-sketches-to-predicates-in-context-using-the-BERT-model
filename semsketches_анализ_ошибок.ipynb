Загрузка верных предсказаний организаторов

```python
SKETCHES_DEV = "data/dev/dev.gold"
dev_gold = json.load(open(SKETCHES_DEV))
```

```python
import pandas as pd

CONTEXTS_DEV = "data/dev/contexts_dev.data"
contexts_dev_df = pd.read_json(CONTEXTS_DEV).transpose()
```

Pymorphy2 для лемматизации предикатов в предложении

```python
!pip install pymorphy2
```

```python
from pymorphy2 import MorphAnalyzer

morph = MorphAnalyzer()

def tokenize_word(word: str, morph=morph) -> str:
  normalized = morph.parse(word)[0].normal_form
  return normalized
```

```python
contexts_dev_df['normalised_instance'] = contexts_dev_df.instance.apply(tokenize_word)
```

```python
dev_gold = {sketch:sent for sent, sketch in dev_gold.items()}
dev_ours = {sketch:sent for sent, sketch in dev_ours.items()}
```

Напечатаем для каждого предложения попарно наше предсказание и верное предсказание задачи

```python
with open('sketch_mistakes.txt', 'w'):
  for sk in dev_ours.keys():
    print(sk)
    print(contexts_dev_df.loc[dev_gold[sk]]['normalised_instance'])
    print(contexts_dev_df.loc[dev_ours[sk]]['normalised_instance'])
    print('---------------------------------')
```
